```python
member_dic = {"侍太郎": 35, "侍花子": 18, "侍一郎": 10}

class Human:
  def __init__(self, name: str, age: int):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20:
      print(f"{self.name}は、大人である")
    else:
      print(f"{self.name}は、大人でない")

for name in member_dic:
  age = member_dic[name]
  human = Human(name, age)
  human.check_adult()
```
