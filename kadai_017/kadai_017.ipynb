```python
member_dic = {}
name_list =[]

class Human:
  def __init__(self, name: str, age: int):
    # 名前(name)と年齢(age)の属性を持つ
    self.name = name
    self.age = age
    # リスト・ディクショナリに追加
    name_list.append(self.name)
    member_dic[self.name] = self.age

  # 以下の条件で標準出力(print)する
  #  ageが20以上の場合に大人である
  #  そうでない場合に大人でない
  def check_adult(self, name: str, age: int):
    if age >= 20:
      print(f"{name}は、大人である")
    else:
      print(f"{name}は、大人でない")

# インスタンスを複数生成
human = Human("侍太郎", 35)
human = Human("侍一郎", 18)
human = Human("侍花子", 20)


# リストの要素数分だけcheck_adultメソッドを呼び出す
for name in name_list:
  human.check_adult(name, member_dic[name])
```
